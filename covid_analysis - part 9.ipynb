# Loading Dataset

```python
import pandas as pd
import numpy as np

df = pd.read_csv('country_vaccination.csv')
```

```python
df.head()
```

### 1. Country with least total Vaccinations

```python
for i in df.sort_values( by = 'total_vaccinations', ascending = True).values [:1]:
    print('Country with least total vaccinations : '+ str(i[0])+ ' - ' + str(i[3]))
```

### 2. Country with the most total Vaccinations

```python
for i in df.sort_values( by = 'total_vaccinations', ascending = False).values [:1]:
    print('Country with the most total vaccinations : '+ str(i[0])+ ' - ' + str(i[3]))
```
