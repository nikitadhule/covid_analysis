# Loading Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('country_vaccination.csv')
```

```python
df
```

# Data Visualization 

## 1. Number of Total Vaccinations in India in Year 2020, 2021 and 2022


```python
data = df.values
```

```python
temp = []
for i in data:
    if(i[0] == 'India'):
        temp.append([i[15],i[18],i[3]])
        
df_pr = pd.DataFrame(temp , columns = ['year_','month','total_vacc'])
df_pr
```

```python
x1 = df_pr['year_']
x2 = df_pr['total_vacc']

plt.plot(x1,x2, color = 'b')

plt.title('Number of Total Vaccinations in India in Year 2020, 2021 and 2022')
plt.xlabel("Year")
plt.ylabel("Number of total vaccinations")


plt.show()
```

## 2. Comparision of number of total vaccinations in year 2020 of India and USA

```python
india_usa = [df[df.country == 'India'] , df[df.country == 'United States']]
result = pd.concat(india_usa)
result[result['year'] == 2020]
```

```python
x1 = result.sort_values(by = 'total_vaccinations' ,ascending= False)['country'].values
x2 = result.sort_values(by = 'total_vaccinations' ,ascending= False)['total_vaccinations'].values

plt.bar(x1,x2, color = 'b')

plt.title('Number of Total Vaccinations in Year 2020 of India and USA')
plt.xlabel("country")
plt.ylabel("Number of total vaccinations")
plt.show()
```

## 3. Comparision of number of total vaccinations in year 2021 of India and China

```python
india_china = [df[df.country == 'China'] , df[df.country == 'India']]
result_ = pd.concat(india_china)
result_[result_['year'] == 2021]
```

```python
x1 = result_['country']
x2 = result_['total_vaccinations']

plt.bar(x1,x2, color = 'r', )
plt.title('Number of Total Vaccinations in Year 2021 of India and China')
plt.xlabel("country")
plt.ylabel("Number of total vaccinations")
plt.show()
```

## 4. Number of Vaccinations in each month in India in the year 2021

```python
data = df.values
temp = []
for i in data:
    if(i[15] == 2021):
        temp.append([i[18], i[3]])
        
df_ = pd.DataFrame( temp, columns = ['month','Total_vacc'])
df_.groupby('month').sum()['Total_vacc']
```

```python
x1 = df_.sort_values(by = 'Total_vacc',ascending = True)['month']
x2 = df_.sort_values(by = 'Total_vacc',ascending = True)['Total_vacc']

fig, ax = plt.subplots(figsize = (8,4))
ax.bar(x1,x2)

plt.title('Number of vaccination in each month in India in the year 2021')
plt.xlabel('Month')
plt.ylabel('Number of vaccination')
plt.show()
```

## 5. The month Which has the most number of total vaccinations in India in 2021

```python
df_.groupby('month').sum().plot(kind = 'pie', y = 'Total_vacc',autopct = '%1.0f%%')
plt.show()
```

### Conclusion : Dec has the most number of total vaccinations in India in the year 2021
