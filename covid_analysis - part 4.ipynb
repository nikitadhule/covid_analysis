# Loading Dataset

```python
import pandas as pd
import numpy as np

from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('country_vaccination.csv')
```

```python
df
```

## (A)Dealing with null values

```python
df.isnull().sum()
```

```python
imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')

s1 =  np.reshape(df['total_vaccinations'].values, (86512,1))                     # s1 for column total_vaccinations   
imputer.fit(s1)
s1 = imputer.transform(s1)
df['total_vaccinations'] = s1

s2 =  np.reshape(df['people_vaccinated'].values, (86512,1))                      # s2 for column people_vaccinated   
imputer.fit(s2)
s2 = imputer.transform(s2)
df['people_vaccinated'] = s2

s3 =  np.reshape(df['people_fully_vaccinated'].values, (86512,1))                # s3 for column people_fully_vaccinated   
imputer.fit(s3)
s3 = imputer.transform(s3)
df['people_fully_vaccinated'] = s3

s4 =  np.reshape(df['daily_vaccinations_raw'].values, (86512,1))                 # s4 for column daily_vaccinations_raw   
imputer.fit(s4)
s4 = imputer.transform(s4)
df['daily_vaccinations_raw'] = s4

s5 =  np.reshape(df['daily_vaccinations'].values, (86512,1))                     # s5 for column daily_vaccinations   
imputer.fit(s5)
s5 = imputer.transform(s5)
df['daily_vaccinations'] = s5

s6 =  np.reshape(df['total_vaccinations_per_hundred'].values, (86512,1))         # s6 for column total_vaccinations_per_hundred   
imputer.fit(s6)
s6 = imputer.transform(s6)
df['total_vaccinations_per_hundred'] = s6

s7 =  np.reshape(df['people_vaccinated_per_hundred'].values, (86512,1))          # s7 for column people_vaccinated_per_hundred   
imputer.fit(s7)
s7 = imputer.transform(s7)
df['people_vaccinated_per_hundred'] = s7

s8 =  np.reshape(df['people_fully_vaccinated_per_hundred'].values, (86512,1))    # s8 for column people_fully_vaccinated_per_hundred   
imputer.fit(s8)
s8 = imputer.transform(s8)
df['people_fully_vaccinated_per_hundred'] = s8

s9 =  np.reshape(df['daily_vaccinations_per_million'].values, (86512,1))             # s9 for column daily_vaccinations_per_million   
imputer.fit(s9)
s9 = imputer.transform(s1)
df['daily_vaccinations_per_million'] = s9
```

```python
df.to_csv('country_vaccination.csv',index= False)
df.head()
```
